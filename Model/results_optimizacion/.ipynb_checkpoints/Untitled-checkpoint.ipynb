```python
def convert_individual_to_vecinos(individual,dist_crimen,n=n):
        
        vecinos=[[i,c] for i,c in zip(range(n),dist_crimen)]
        #print(vecinos)
        k=0
        for i in range(n-1):    
            I=i*n-k
            F=(i+1)*(n-1)-k
            k+=(i+1)
            #print(individual[I:F])
            v=np.where(individual[I:F])[0]+i+1
            
            for j in v:
                vecinos[i].append(j)
                vecinos[j].append(i)
        return vecinos
```
