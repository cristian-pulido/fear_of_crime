```python
from AG_basic import *
```

```python
def Rastrigin(individual):
    import math
    lista=[int("".join(map(str, individual[i*10:10*(i+1)].tolist())),2)/100-5.12 for i in range(12)]
    A=10
    return (sum([(A+x**2 - A * np.cos(2 * math.pi * x)) for x in lista]))
```

```python
AG_1=AG(nombre="suma",
        persons=16,
        individuals=1000,
        mode_initial_pop=pop_ini_uniform,#pop_ini_uniform,
        n_generations=300,
        p_crossover=0.7,
        p_mutation=None,
        crossover_func=ruleta,
        generacional=False,
        fitness_func=Rastrigin
       )
```

```python
AG_1.info()
```

```python
AG_1.run_alone(path_dir="sum/slower",console=False,plots=3)
```

```python
a=AG_1.current_population[np.argmin(AG_1.evaluate_pop(AG_1.current_population,offspring=True))]
lista=[int("".join(map(str, a[i*10:10*(i+1)].tolist())),2)/100-5.12 for i in range(12)]
lista
```
