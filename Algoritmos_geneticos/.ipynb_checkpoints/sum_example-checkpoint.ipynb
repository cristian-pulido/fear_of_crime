```python
def Rastrigin(individual):
    import math
    lista=[int("".join(map(str, individual[i*10:10*(i+1)].tolist())),2)/100-5.12 for i in range(9)]
    A=10
    return -sum([(x**2 - A * np.cos(2 * math.pi * x)) for x in lista])
```

```python
Rastrigin(AG_1.current_population[np.random.choice(AG_1.individuals)])
```
