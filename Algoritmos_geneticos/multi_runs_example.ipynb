```python
def Rastrigin(individual):
    import math
    lista=[int("".join(map(str, individual[i*10:10*(i+1)].tolist())),2)/100-5.12 for i in range(12)]
    A=10
    return (sum([(A+x**2 - A * np.cos(2 * math.pi * x)) for x in lista])+0.01)
```
